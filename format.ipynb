```python
# Import Libraries
import pandas as pd
import glob
import os
```

```python
data_path = r"C:\Users\Abhishek\Desktop\quantium-starter-repo\data"

# Get all CSV files in the folder
csv_files = glob.glob(os.path.join(data_path, "*.csv"))
```

```python
print(csv_files)
```

```python
# Check if any CSV files are found
if not csv_files:
    print("No CSV files found in the folder.")
else:
    # Load each CSV into a DataFrame
    df_list = [pd.read_csv(file) for file in csv_files]

    # Combine all DataFrames into one
    data = pd.concat(df_list, ignore_index=True)

    # Optional: preview the combined DataFrame
    print(data.head())
    print(f"Total rows: {len(data)}")
    
```

```python
data = data[data["product"] == "pink morsel"]
```

```python
data["Sales"] = data["quantity"] * data["price"]
```

```python
formatted = data[["Sales", "date", "region"]].copy()
formatted.rename(columns={"date": "Date", "region": "Region"}, inplace=True)
```

```python
formatted.to_csv("formatted_sales.csv", index=False)
```
